```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model # Build and train machine learning models with Scikit-learn
```

```python
#Simple linear regression
#create a sample dataframe
df = pd.DataFrame({
    'area': [2500, 3000, 3200, 3600, 4000,4500],
    'price': [55000, 56000, 61000, 66000, 70000,75000]
})
```

```python
df
```

```python
#scatter Plot create
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price) #Creates a scatter plot of area vs price using the df DataFrame
plt.show()
```

```python
#drop price columns
new_df = df.drop('price', axis='columns')
new_df
```

```python
model = linear_model.LinearRegression()  
model.fit(new_df,df.price) # creating a new linear regression model and training it on the input features new_df and the target variable df.price.
```

```python
model.predict([[5000]]) #expect 2D array
```

```python
model.coef_ #property of linear regression
```

```python
model.intercept_
```
